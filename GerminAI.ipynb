```python
#adicionar o agente de diagnóstico
import re
import requests

def geolocalizar_diagnostico_completo(local_input):
    def is_coord(texto):
        return re.match(r"^-?\d{1,3}\.\d+,\s*-?\d{1,3}\.\d+$", texto.strip())

    def get_coords_from_text(text):
        url = f"https://nominatim.openstreetmap.org/search?q={text}&format=json"
        headers = {"User-Agent": "Mozilla/5.0"}

        try:
            response = requests.get(url, headers=headers, timeout=10)

            if response.status_code != 200 or not response.text.strip():
                print(f"❌ Erro ou resposta vazia da API Nominatim: {response.status_code}")
                return None, None, None

            data = response.json()
            if not data:
                print("⚠️ Nenhuma coordenada encontrada para:", text)
                return None, None, None

            return data[0]["lat"], data[0]["lon"], data[0]["display_name"]

        except Exception as e:
            print("⚠️ Erro ao buscar coordenadas:", str(e))
            return None, None, None

    def reverse_lookup(lat, lon):
        url = f"https://nominatim.openstreetmap.org/reverse?lat={lat}&lon={lon}&format=json"
        headers = {"User-Agent": "Mozilla/5.0"}
        try:
            r = requests.get(url, headers=headers).json()
            return r.get("display_name", ""), r.get("address", {})
        except:
            return "", {}

    def consultar_bioma_ibge(cidade_nome):
        url = f"https://servicodados.ibge.gov.br/api/v1/localidades/municipios"
        try:
            r = requests.get(url, timeout=10).json()
            for cidade in r:
                if cidade_nome.lower() in cidade["nome"].lower():
                    microrregiao = cidade["microrregiao"]["nome"]
                    regiao = cidade["microrregiao"]["mesorregiao"]["UF"]["regiao"]["nome"]
                    estado = cidade["microrregiao"]["mesorregiao"]["UF"]["nome"]
                    return estado, microrregiao, regiao
        except:
            pass
        return None, None, None

    # Identificar tipo de entrada e obter lat/lon
    if is_coord(local_input):
        lat, lon = local_input.split(",")
        local_formatado, endereco = reverse_lookup(lat.strip(), lon.strip())
    else:
        lat, lon, local_formatado = get_coords_from_text(local_input)
        if lat and lon:
            local_formatado, endereco = reverse_lookup(lat, lon)
        else:
            return "❌ Localização não reconhecida automaticamente.", None, None

    pais = endereco.get("country", "desconhecido")
    estado = endereco.get("state", "")
    cidade = endereco.get("city", endereco.get("town", endereco.get("village", "desconhecida")))

    # Dados do IBGE se for Brasil
    microrregiao, regiao_geo, estado_ibge = None, None, None
    if pais.lower() == "brazil":
        estado_ibge, microrregiao, regiao_geo = consultar_bioma_ibge(cidade)

    # Bioma estimado
    if pais.lower() == "brazil":
        if estado.lower() in ["amazonas", "pará", "acre", "roraima", "amapá", "rondônia"]:
            bioma = "Amazônia"
            solo = "latossolo argiloso"
        elif estado.lower() in ["goiás", "mato grosso", "tocantins", "distrito federal"]:
            bioma = "Cerrado"
            solo = "arenoso com baixa matéria orgânica"
        elif estado.lower() in ["bahia", "ceará", "pernambuco", "sergipe", "piaui"]:
            bioma = "Caatinga"
            solo = "raso e pedregoso"
        elif estado.lower() in ["rs", "sc", "paraná"]:
            bioma = "Mata Atlântica"
            solo = "latossolo vermelho com matéria orgânica"
        elif estado.lower() in ["mato grosso do sul"]:
            bioma = "Pantanal"
            solo = "hidromórfico com alta matéria orgânica"
        else:
            bioma = "Bioma brasileiro estimado"
            solo = "solo regional médio"
    else:
        bioma = "Vegetação estimada por latitude"
        if float(lat) < -30:
            solo = "argiloso ou frio"
        elif float(lat) > 40:
            solo = "temperado seco ou arenoso"
        else:
            solo = "solo tropical ou subtropical médio"

    resposta = f"""🌍 Localização reconhecida: {cidade}, {estado}, {pais}
🗺️ Bioma estimado: {bioma}
🧱 Tipo de solo provável: {solo}
🌐 Coordenadas: {lat}, {lon}
"""
    if microrregiao:
        resposta += f"🏞️ Microrregião IBGE: {microrregiao}\n"
    if regiao_geo:
        resposta += f"🧭 Região geográfica IBGE: {regiao_geo}\n"

    resposta += "📌 Fonte: Nominatim (OpenStreetMap), IBGE API"

    return resposta, float(lat), float(lon)
```

```python
#Agente de clima
def buscar_clima(latitude, longitude):
    url = f"https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&current_weather=true"
    try:
        response = requests.get(url, timeout=10)

        if response.status_code != 200:
            print("⚠️ Erro ao consultar clima:", response.status_code)
            return {"temperature": "N/A", "windspeed": "N/A"}

        clima = response.json()

        if "current_weather" in clima:
            return clima["current_weather"]
        else:
            print("⚠️ Resposta inesperada da API de clima.")
            return {"temperature": "N/A", "windspeed": "N/A"}

    except Exception as e:
        print("⚠️ Erro ao buscar clima:", str(e))
        return {"temperature": "N/A", "windspeed": "N/A"}

```

```python
# Agente Google Scholar (artigos científicos)
from bs4 import BeautifulSoup
import requests

def buscar_artigos_scholar(pergunta):
    query = pergunta.replace(" ", "+")
    url = f"https://scholar.google.com/scholar?q={query}+agricultura+sintrópica"
    headers = {"User-Agent": "Mozilla/5.0"}
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "html.parser")
    resultados = soup.select(".gs_rt a")
    links = [r["href"] for r in resultados[:3] if r.has_attr("href")]
    return links
```

```python
#Agente Site do Götsch (via Google Search)
def buscar_no_site_gotsch(query):
    url = f"https://www.google.com/search?q=site:agendagotsch.com/pt {query}"
    headers = {"User-Agent": "Mozilla/5.0"}
    resp = requests.get(url, headers=headers)
    soup = BeautifulSoup(resp.text, "html.parser")
    resultados = soup.select("div.yuRUbf > a")
    return [r["href"] for r in resultados[:3]]
```

```python
#Autenticação com Google Calendar
#Função de agendamento
import os
import datetime
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build

# Escopo de permissão para editar a agenda
SCOPES = ['https://www.googleapis.com/auth/calendar']

# Caminho do seu arquivo .json de credenciais (exatamente como está no Colab)
CREDENTIALS_PATH = "/content/client_secret.json"
TOKEN_PATH = '/content/token.json'

def autenticar_google_calendar():
    creds = None
    if os.path.exists(TOKEN_PATH):
        creds = Credentials.from_authorized_user_file(TOKEN_PATH, SCOPES)
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file(CREDENTIALS_PATH, SCOPES)
            creds = flow.run_local_server(port=0, open_browser=False)
        with open(TOKEN_PATH, 'w') as token:
            token.write(creds.to_json())
    return build('calendar', 'v3', credentials=creds)


```

```python
# Função para agendar evento
def agendar_no_google_calendar(titulo, descricao, dias_a_partir_de_hoje):
    service = autenticar_google_calendar()

    hoje = datetime.datetime.utcnow()
    inicio = hoje + datetime.timedelta(days=dias_a_partir_de_hoje)
    fim = inicio + datetime.timedelta(hours=1)

    evento = {
        'summary': titulo,
        'description': descricao,
        'start': {'dateTime': inicio.isoformat() + 'Z'},
        'end': {'dateTime': fim.isoformat() + 'Z'},
    }

    evento_criado = service.events().insert(calendarId='primary', body=evento).execute()
    print(f"✅ Evento adicionado: {evento_criado.get('htmlLink')}")
```

```python
def agendar_eventos_dinamicamente(eventos):
    print("\n Deseja adicionar algum desses eventos à sua agenda?")
    for i, evento in enumerate(eventos, 1):
        print(f"{i}. {evento['titulo']} (em {evento['dias']} dias)")

    selecao = input("\nDigite os números dos eventos separados por vírgula (ex: 1,3,5), ou pressione Enter para pular: ")

    if selecao.strip() == "":
        print("Nenhum evento será adicionado.")
        return

    indices = [int(i.strip()) - 1 for i in selecao.split(",") if i.strip().isdigit()]
    for idx in indices:
        if 0 <= idx < len(eventos):
            ev = eventos[idx]
            agendar_no_google_calendar(ev['titulo'], ev['descricao'], ev['dias'])
```
